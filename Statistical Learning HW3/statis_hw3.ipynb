```
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt 
digits = load_digits()
print(digits.data.shape)
for i in range(10):  
  plt.gray() 
  plt.matshow(digits.images[i]) 
  plt.show() 
```

```
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=.25,shuffle=True)
####################classifiers=[QDA(),LDA(),LogisticRegression()]###################
digits.target
```

###############**Qudratic Discriminant Analysis Classifier**#############

```
import matplotlib.pyplot as plt
import seaborn as sns
model=QuadraticDiscriminantAnalysis()
model.fit(x_train,y_train)
print('accuracy for QDA is',model.score(x_test,y_test))
y_pred=model.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
plt.figure(figsize=(10,7))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True, fmt="d",linewidths=.5)
```

#########**Linear Discriminant Analysis Classifier**#########

```
model=LinearDiscriminantAnalysis()
model.fit(x_train,y_train)
print('accuracy for LDA is',model.score(x_test,y_test))
y_pred=model.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
plt.figure(figsize=(10,7))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True, fmt="d",linewidths=.5)
```

###############**Logistic Regression Classifier**############

```
model=LogisticRegression(penalty='l2',solver='lbfgs',multi_class='multinomial',max_iter=10000)
model.fit(x_train,y_train)
print('accuracy for Logistic Regression is',model.score(x_test,y_test))
y_pred=model.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
plt.figure(figsize=(10,7))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True, fmt="d",linewidths=.5)
```
